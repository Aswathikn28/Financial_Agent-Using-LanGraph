## AGENT CREATION TASKS

### Financial Analysis Multi-Agent System

* Build a system that:

   1. Fetches financial data

   2. Performs calculations

   3. Analyzes risk

   4. Generates investment advice

* Required Agents:

   1. Planner Agent: Identify required financial metric

   2. Data Agent: Calls financial API tool,Stores numeric value
   
   3. Analyzer Agent: Computes ratios or projections, Updates state
   
   4. Risk Agent: Classifies risk level (Low/Medium/High)
   
   5. Advisor Agent: Generates final investment suggestion

```python
from typing import TypedDict, Annotated
import yfinance as yf
from langchain_openai import ChatOpenAI
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.messages import HumanMessage
from langchain_core.tools import tool
from langgraph.graph import StateGraph, END
from langgraph.graph.message import add_messages
from langgraph.prebuilt import ToolNode
```

```python
# GEMINI LLM

import os
os.environ["GOOGLE_API_KEY"] = "YOUR_API_KEY"

llm = ChatGoogleGenerativeAI(
     model="models/gemini-flash-lite-latest",  
    temperature=0
)
```

```python
import google.generativeai as genai
import os

genai.configure(api_key=os.environ["GOOGLE_API_KEY"])

for m in genai.list_models():
    print(m.name)
```

```python
# STATE
class FinancialState(TypedDict):
    messages: Annotated[list, add_messages]
    company: str
    metrics_required: list
    financial_data: dict
    analysis: dict
    risk_level: str
    advice: str
```

```python
# -----------------------------
# TOOL: Financial API
# -----------------------------
@tool
def fetch_financial_data(company: str) -> dict:
    """Fetch financial metrics for a company ticker."""
    stock = yf.Ticker(company)
    info = stock.info
    
    return {
        "current_price": info.get("currentPrice", 0),
        "market_cap": info.get("marketCap", 0),
        "pe_ratio": info.get("trailingPE", 0),
        "revenue_growth": info.get("revenueGrowth", 0),
        "debt_to_equity": info.get("debtToEquity", 0)
    }
```

```python
# -----------------------------
# PLANNER AGENT (LLM)
# -----------------------------
def planner_node(state: FinancialState):
    company = state["company"]

    prompt = f"""
    You are a financial planning agent.
    Identify important financial metrics needed to evaluate {company}.
    Return them as a Python list.
    """

    response = llm.invoke(prompt)

    return {
        "metrics_required": [
            "current_price",
            "market_cap",
            "pe_ratio",
            "revenue_growth",
            "debt_to_equity"
        ],
        "messages": [response]
    }
```

```python
# -----------------------------
# DATA AGENT
# -----------------------------
def data_node(state: FinancialState):
    company = state["company"]
    data = fetch_financial_data.invoke(company)

    return {"financial_data": data}
```

```python
# -----------------------------
# ANALYZER AGENT (LLM reasoning)
# -----------------------------
def analyzer_node(state: FinancialState):
    data = state["financial_data"]

    prompt = f"""
    Analyze the following financial data:
    {data}

    Evaluate financial strength and give:
    - financial_score (1 to 3)
    - short explanation
    Return JSON format.
    """

    response = llm.invoke(prompt)

    return {
        "analysis": {
            "raw_analysis": response.content
        }
    }
```

```python
# -----------------------------
# RISK AGENT (Rule-Based)
# -----------------------------
def risk_node(state: FinancialState):

    company = state["company"]
    analysis = state["analysis"]

    debt_to_equity = analysis.get("debt_to_equity", 0)
    interest_coverage = analysis.get("interest_coverage", 0)
    free_cash_flow = analysis.get("free_cash_flow", 0)

    bluechip_companies = ["AAPL", "MSFT", "GOOGL", "AMZN"]

    # Risk Logic
    if debt_to_equity > 2:
        if company in bluechip_companies and interest_coverage > 5:
            risk = "Medium"
        else:
            risk = "High"
    elif debt_to_equity <= 2 and free_cash_flow > 0:
        risk = "Low"
    else:
        risk = "Medium"

    return {
        "risk_level": risk
    }
```

```python
# -----------------------------
# ADVISOR AGENT
# -----------------------------
def advisor_node(state: FinancialState):
    risk = state["risk_level"]
    company = state["company"]

    prompt = f"""
    Company: {company}
    Risk Level: {risk}

    Provide a final investment recommendation.
    """

    response = llm.invoke(prompt)

    return {"advice": response.content}
```

```python
# -----------------------------
# BUILD GRAPH
# -----------------------------
builder = StateGraph(FinancialState)

builder.add_node("planner", planner_node)
builder.add_node("data", data_node)
builder.add_node("analyzer", analyzer_node)
builder.add_node("risk", risk_node)
builder.add_node("advisor", advisor_node)

builder.set_entry_point("planner")

builder.add_edge("planner", "data")
builder.add_edge("data", "analyzer")
builder.add_edge("analyzer", "risk")
builder.add_edge("risk", "advisor")
builder.add_edge("advisor", END)

graph = builder.compile()
```

```python
# -----------------------------
# RUN SYSTEM
# -----------------------------
initial_state = {
    "messages": [],
    "company": "ORCL",
    "metrics_required": [],
    "financial_data": {},
    "analysis": {},
    "risk_level": "",
    "advice": ""
}
result = graph.invoke(initial_state)
```
